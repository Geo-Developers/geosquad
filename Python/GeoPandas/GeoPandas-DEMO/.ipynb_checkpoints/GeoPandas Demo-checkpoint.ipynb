### 1. Import Modules
---
First thing we do is import the libraries that we are going to use during the demo. In this case, we use `matplotlib`, `pandas`, `geopandas` and `shapely`.

```python
%matplotlib inline
import pandas as pd
import geopandas
from shapely.geometry import Point, Polygon
pd.options.display.max_rows = 10
```

### 2. Read and write GIS file formats
---
GeoPandas can read and write many different GID file formats by relying on the `fiona` library (which is an interface to GDAL/OGR).

For example, let's start by reading a GeoJSON file about the emplacement stations in Paris (vélib, https://opendata.paris.fr/explore/dataset/velib-emplacement-des-stations/) and administrative districts in Paris (https://opendata.paris.fr/explore/dataset/quartier_paris/):

```python
stations = geopandas.read_file("data/velib-emplacement-des-stations.geojson")
stations[['name', 'capacity', 'geometry']]
```

```python
quartiers = geopandas.read_file("data/quartier_paris.geojson")
quartiers[['l_qu', 'geometry']]
```

### 3. What are GeoDataFrames and GeoSeries?
---
* **GeoDataFrame** is a Data Frame which has a 'geometry' column. The `.geometry` attribute returns a GeoSeries (the column name itself is not necessarily 'geometry').
* **GeoSeries** is a `Series` that holds (shapely) geometry objects (Points, LineStrings, Polygons, ...).

We can check if the data is a GeoDataFrame or GeoSeries by using the [type()](https://docs.python.org/3/library/functions.html#type) method:

```python
type(quartiers)
```

```python
quartiers.geometry
```

```python
type(quartiers.geometry)
```

```python
type(quartiers.geometry[0])
```

#### 3.1. It's still a DataFrame
---
We still have our pandas functionality that we can use on the geospatial dataset (e.g. boolean filtering).

```python
quartiers = quartiers[quartiers['c_qu'] <= 30].copy()
```

An histogram showing the distribution.

```python
quartiers['surface'].hist()
```

#### 3.2. But now has access to spacial predicates and operations
---
For example, we can acces the area of the different quartiers:

```python
quartiers.geometry.area
```

Or calculate the distance from a certain point to all the quartiers.

I used geopy to localize the Notre Dame:
    >>> from.geopy.geocoders import Nominatim
    >>> geolocator = Nominatim()
    >>> location = geolocator.geocode("Notre Dame Paris")
    >>> notre_dame = Point(location.longitude, location.latitude)
    >>> print(notre_dame)
    POINT (2.35005149954546 48.85293695)

```python
notre_dame = Point(2.35005149954546, 48.85293695)
```

```python
stations.geometry.distance(notre_dame)
```

Or check in which quartier the Notre Dame is located:

```python
quartiers.contains(notre_dame)
```

```python
quartiers[quartiers.contains(notre_dame)]
```

Many spatial predicates and operations:

* **Predicates**: contains, covers, crosses, intersects, equals, overlaps, touches, within...
* **Operations**: difference, intersection, union, buffer, distance, length, area, interpolate, project, relate...

### 4. Visualize your geospatial data
---
GeoPandas allows you to quickly visualize your geometries:

```python
quartiers.plot(figsize=(10, 10))
```

```python
ax = quartiers.plot(figsize=(10, 10), cmap='tab10', alpha=0.5)
ax.set_axis_off()
```

```python
ax = quartiers.plot(figsize=(10, 10), edgecolor='k', facecolor='r', alpha=0.5, linewidth=2)
ax.set_axis_off()
```

```python
stations.plot(markersize=5, figsize=(10, 10))
```

It would be nice to add the street network to the figure above. Therefore, I downloaded, using the `osmnx` package, the openstreetmap network for Paris, and saved it as a shapefile:

```python
import osmnx as ox

G = ox.graph_from_place('Paris, France', network_type='drive')

ox.save_graph_shapefile(G, 'openstreetmap_paris', folder='./data')
```

```python
streets = geopandas.read_file("data/openstreetmap_paris/edges/")
streets.head()
```

Then we can combine plots to show both DataFrames:

```python
ax = streets.plot(linewidth=0.5, color='grey', figsize=(10, 10))
stations.plot(ax=ax, markersize=20)
ax.set(xlim=(2.25, 2.42), ylim=(48.815, 48.905))
ax.set_axis_off()
```

#### 4.1. Color the plot based on values
---
By specifying the column keyword, we can color the geometries based on the values of that column:

```python
ax = streets.plot(linewidth=0.5, color='grey', figsize=(10, 10))
quartiers.plot(ax=ax, column='l_qu', cmap='tab10', legend=True)
stations.plot(ax=ax, column='capacity', markersize=50, legend=True)

ax.set(xlim=(2.25, 2.42), ylim=(48.815, 48.905))
ax.set_axis_off()
```

Or based on a categorical values:

```python
ax = quartiers.plot(column='n_sq_ar', categorical=True, cmap='tab20', figsize=(10, 10))
ax.set_axis_off()
```

#### 4.1. Spatial Joins
---
GeoPandas includes functionallity to perform spatial joins and overlays (`sjoin` and `overlay`). 

Here, let's do a spatial join to determine the quartier on which each station is located.

```python
stations = geopandas.sjoin(stations, quartiers[['l_qu', 'geometry']].copy(), op='within')
```

```python
stations.head()
```

Now lets do a merge to add data from stations DataFrame to quartiers DataFrame.

```python
counts = stations.groupby('l_qu').size()
```

```python
quartiers = quartiers.merge(counts.reset_index(name='number_stations'))
```

```python
quartiers.head()
```

This creates a column that counts how many points (stations) are in that area (quartier).

Now lets create a plot that displays the number of stations in the quartiers.

```python
ax = quartiers.plot(column='number_stations', figsize=(10, 10), legend=True)
ax.set_axis_off()
```

Then we can create relative values to display a more intuitive plot.

```python
quartiers['number_stations_relative'] = (quartiers['number_stations'] / quartiers.geometry.area)
```

```python
ax = quartiers.plot(column='number_stations_relative', figsize=(10, 10))
ax.set_axis_off()
```

### 5. Coordinate reference systems
---
A GeoDataFrame or GeoSeries has a `.crs` attribute which holds (optionally) a description of the coordinate reference system of the geometries:

```python
stations.crs
```

```python
stations.geometry.head()
```

We can convert this to another reference system using the `to_crs` function. For example, let's convert it to the UTM 31 zone reference system (http://epsg.io/32631) which unit is meters.

```python
stations2 = stations.to_crs(epsg=32631)
# or
# stations,to_crs("+proj=utm +zone=31 +ellps=WGS84)
```

```python
stations2.geometry.head()
```

```python
quartiers2 = quartiers.to_crs(epsg=32631)
```

```python
(quartiers2.geometry.area / 1000**2)
```

### 6. Interactive Leaflet.js maps with folium library
---
Here's an example of how to create a choropleth map using folium library.

```python
import folium
```

```python
m = folium.Map([48.8566, 2.3429], zoom_start=12, tiles="OpenStreetMap")
m.choropleth(geo_str=quartiers.to_json(), data=quartiers, columns=['objectid', 'surface'], key_on='feature.properties.objectid', fill_color='BuGn')
m
```
