```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataFrame = pd.read_csv('data/battles.csv')
```

```python
dataFrame.dtypes
```

```python
dataFrame.head(10)
```

```python
sns.set_style('whitegrid')
example_one = sns.barplot(x='year', y='major_death', data=dataFrame)
```

```python
example_two = sns.factorplot(x='attacker_size', y='defender_size', hue='attacker_outcome', data=dataFrame, size=6, kind='bar', palette='colorblind')
plt.xticks(rotation=60)
plt.rcParams['xtick.labelsize'] = 9
```

```python
import numpy as np

sns.set(style="white")

rs = np.random.RandomState(10)

mean = (2,2)

cov = [(1,.5),(.5,1)]

x1, x2 = rs.multivariate_normal(mean, cov, 500).T

x1 = pd.Series(x1, name="Example_X")
x2 = pd.Series(x2, name="Example_Y")

example_three = sns.jointplot(x1, x2, kind="kde", size=7, space=0, color="g")
```

```python
example_ex = sns.lmplot(x="year", y="battle_number", data=dataFrame, fit_reg=False, hue="battle_type", size=6, )
```
