### 6.1. Renaiming
---
You can use `pandas` renaming utility functions to change the names of columns or indexes by using the `rename` function, which lets you rename index and/or column names. For example, to change a column name we would do:

```python
battles_got.rename(columns={'name': 'battle_name'})
```

`rename` lets you rename index or column values by specifying a `index` or `column` keyword parameter.

You can use the `rename_axis` method to change the attribute name of the row index or column index.

```python
battles_got.rename_axis("battles", axis='rows').rename_axis("fields", axis='columns')
```
